```python
import pandas as pd
import numpy as np
% matplotlib inline
df = pd.read_table('iris.data', header=None)
df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']] = df.iloc[:,0].str.split(",", expand=True)
```

```python
df.head()
```

```python
df = pd.DataFrame.drop(df, axis=1, columns=0)
```

```python
df.head()
```

```python
# There's gotta be a better way!
df['sepal_length'] = df['sepal_length'].astype(float)
df['sepal_width'] = df['sepal_width'].astype(float)
df['petal_length'] = df['petal_length'].astype(float)
df['petal_width'] = df['petal_width'].astype(float)
```

```python
df.info()
```

```python
df.plot(x='sepal_length', y='sepal_width');
```
